# classes

```python
class Registre:
    def __init__(self, nom, telf, edat):
        self.nom = nom
        self.telf = telf
        self.edat = edat

    def __eq__(self, unaltra):
        """ compara dos registres """
        if not isinstance(unaltra, Registre):
            return False
        return self.nom == unaltra.nom

    def __call__(self):
        print(f"{self.nom=}")


r1 = Registre("Alex", "66644433",21)
r2 = Registre("Alex", "66644433",21)

help(r1)

r1 == r2

r1()
```
